# EMDB Organism Fetch

This notebook imports the helper functions from `fetch_emdb_organisms.py`, pulls the full EMDB organism list via the paged search API, and writes one row per NCBI taxonomy code to CSV. Organism names are only kept when the EMDB mapping is consistent enough to trust; otherwise the name is left blank.

```python
from pathlib import Path

from fetch_emdb_organisms import collect_organisms, write_output

OUTPUT_PATH = Path("all_emdb_ncbi_taxonomy_codes.csv")
ROWS_PER_PAGE = 10000
TIMEOUT_SECONDS = 60
PAUSE_SECONDS = 0.0
```

```python
organisms = collect_organisms(
    rows=ROWS_PER_PAGE,
    timeout=TIMEOUT_SECONDS,
    pause=PAUSE_SECONDS,
)

write_output(str(OUTPUT_PATH), organisms)

print(f"Wrote {len(organisms)} unique organisms to {OUTPUT_PATH.resolve()}")
```
