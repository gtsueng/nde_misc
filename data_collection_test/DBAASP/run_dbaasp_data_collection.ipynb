# DBAASP Data Collection

This notebook runs the existing DBAASP helper scripts in this folder:

- `fetch_dbaasp_target_organisms.py` to collect all unique target organisms and peptide counts.
- `codex_fetch_dbaasp_peptides_by_target_organism.py` to fetch all peptides for a selected target organism.

Run the setup cell first, then execute either workflow section below.

```python
from pathlib import Path

from codex_fetch_dbaasp_peptides_by_target_organism import collect_peptides, write_output as write_peptide_output
from fetch_dbaasp_target_organisms import (
    OUTPUT_DIR,
    fetch_all_target_organisms,
    get_all_peptide_ids,
    save_results,
)

DBAASP_DIR = Path.cwd()
DBAASP_DIR
```

## Workflow 1: Collect All Target Organisms

```python
TARGET_ORGANISM_OUTPUT = OUTPUT_DIR / "dbaasp_target_organisms.tsv"
TARGET_ORGANISM_OUTPUT
```

```python
peptide_ids = get_all_peptide_ids()
organism_counts = fetch_all_target_organisms(peptide_ids)
save_results(organism_counts, TARGET_ORGANISM_OUTPUT)

print(f"Wrote {len(organism_counts)} target organisms to {TARGET_ORGANISM_OUTPUT.resolve()}")
```

```python
sorted(organism_counts.items(), key=lambda item: item[1], reverse=True)[:20]
```

## Workflow 2: Fetch Peptides For One Target Organism

```python
TARGET_ORGANISM = "Escherichia coli"
PEPTIDE_PAGE_LIMIT = 500
TIMEOUT_SECONDS = 60
PAUSE_SECONDS = 0.0
PEPTIDE_OUTPUT = DBAASP_DIR / "escherichia_coli_peptides.json"

TARGET_ORGANISM, PEPTIDE_OUTPUT
```

```python
peptides = collect_peptides(
    organism=TARGET_ORGANISM,
    limit=PEPTIDE_PAGE_LIMIT,
    timeout=TIMEOUT_SECONDS,
    pause=PAUSE_SECONDS,
)
write_peptide_output(str(PEPTIDE_OUTPUT), TARGET_ORGANISM, peptides)

print(f"Wrote {len(peptides)} peptides for {TARGET_ORGANISM} to {PEPTIDE_OUTPUT.resolve()}")
```
