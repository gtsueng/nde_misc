# MWCCS Substudy Download and Parse

This notebook runs the MWCCS substudy PDF downloader and parser using the `anaconda3/envs/nde` Python environment.


```python
from __future__ import annotations

import json
import subprocess
from pathlib import Path

NOTEBOOK_DIR = Path.cwd().resolve()
if NOTEBOOK_DIR.name != "MWCCS":
    NOTEBOOK_DIR = NOTEBOOK_DIR / "MWCCS"

PYTHON_EXE = Path(r"C:\Users\gtsueng\Anaconda3\envs\nde\python.exe")
DOWNLOAD_SCRIPT = NOTEBOOK_DIR / "download_mwccs_substudy_pdfs.py"
PARSE_SCRIPT = NOTEBOOK_DIR / "parse_mwccs_substudy_pdfs.py"
SUBSTUDY_DIR = NOTEBOOK_DIR / "Substudies"
OUTPUT_JSON = NOTEBOOK_DIR / "mwccs_substudies_datacollections.json"

NOTEBOOK_DIR, PYTHON_EXE
```

```python
def run_command(args: list[str], cwd: Path = NOTEBOOK_DIR) -> subprocess.CompletedProcess[str]:
    result = subprocess.run(
        args,
        cwd=str(cwd),
        capture_output=True,
        text=True,
        encoding="utf-8",
        errors="replace",
        check=False,
    )
    print(result.stdout)
    if result.stderr:
        print(result.stderr)
    if result.returncode != 0:
        raise RuntimeError(f"Command failed with exit code {result.returncode}: {args}")
    return result
```

## Download PDFs

This refreshes the files in `MWCCS/Substudies` from the live MWCCS substudy page.


```python
run_command([str(PYTHON_EXE), str(DOWNLOAD_SCRIPT)])
```

## Parse PDFs into DataCollection JSON

This runs the parser and writes `mwccs_substudies_datacollections.json`.


```python
run_command([str(PYTHON_EXE), str(PARSE_SCRIPT)])
```
